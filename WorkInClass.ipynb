#### Simulating birthdays

1. Pick a random birthday as an integer between $1$ and $365$.

Note that there has been a major update to the numpy family of random function.
Most old teaching materials (including DataCamp) uses these functions:

1. `np.random.permutation`: Return a random permutation of a sequence, or return a permuted range
2. `np.random.integers`: Draw random integers from a given low-to-high range
3. `np.random.choice`: Generates a random sample from a given 1-D array

Whereas the **modern**, **recommended**, and **safer** interface. would be.

`rng = np.random.default_rng()`
1. `rng.permutation(...)`
2. `rng.integers(...)`
3. `rng.choice(...)`

| Purpose            | Old function                            | New (recommended) equivalent      | Key difference                                                |
| ------------------ | --------------------------------------- | --------------------------------- | ------------------------------------------------------------- |
| Random integers    | `np.random.randint(low, high, size)`    | `rng.integers(low, high, size)`   | identical behavior except new one handles ranges more cleanly |
| Random permutation | `np.random.permutation(x)`              | `rng.permutation(x)`              | identical                                                     |
| Random choice      | `np.random.choice(a, size, replace, p)` | `rng.choice(a, size, replace, p)` | identical, but new one defaults to `replace=True` explicitly  |
| Uniform floats     | `np.random.rand()`                      | `rng.random()`                    | identical                                                     |
| Gaussian           | `np.random.normal()`                    | `rng.normal()`                    | identical                                                     |
| Seeding            | `np.random.seed(42)`                    | `rng = np.random.default_rng(42)` | new one avoids global state                                   |
